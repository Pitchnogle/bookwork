```python
knob_weight = 0.5
input = 0.5
goal_pred = 0.8

pred = input * knob_weight
error = (pred - goal_pred) ** 2

print(error)
```

```python
# Hot & Cold Learning

weight = 0.5
input = 0.5
goal_pred = 0.8

step_amount = 0.001

for iteration in range(1101):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    
    print("Error: %.8f\tPrediction: %.8f" % (error, pred))
    
    up_pred = input * (weight + step_amount)
    up_error = (goal_pred - up_pred) ** 2
    
    down_pred = input * (weight - step_amount)
    down_error = (goal_pred - down_pred) ** 2
    
    if (down_error < up_error):
        weight -= step_amount
    
    if (down_error > up_error):
        weight += step_amount
```

```python
# Hot & Cold Learning (with Gradient Descent)

weight = 0.5
input = 0.5
goal_pred = 0.8

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    
    adjust = (pred - goal_pred) * input
    weight = weight - adjust
    
    print("Error: %.8f\tPrediction: %.8f" % (error, pred))
```

```python
weight, goal_pred, input = (0.0, 0.8, 1.1)

for i in range(4):
    print("-----\nWeight: %.8f" % (weight))
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight -= weight_delta
    print("Error: %.8f\tPrediction: %.8f" % (error, pred))
    print("Delta: %.8f\tWeight Delta: %.8f" % (delta, weight_delta))
```

```python
# Alpha in Code

weight = 0.5
goal_pred = 0.8
input = 2
alpha = 0.1 # 0.0 - 1.0

# With an input of 2 this will diverge without alpha!

for i in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    derivative = input * (pred - goal_pred)
    weight -= alpha * derivative
    print("Error: %.8f\tPrediction: %.8f" % (error, pred))
```
