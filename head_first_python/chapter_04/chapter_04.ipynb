```python
# Code Magnets (pg 107)
man = []
other_man = []

try:
    data = open('sketch.txt')

    for each_line in data:
        try:
            (role, line_spoken) = each_line.split(':', 1)
            if role == 'Man':
                man.append(line_spoken)
            elif role == 'Other Man':
                other_man.append(line_spoken)
        except ValueError:
            pass

    data.close()
except IOError:
    print('The data file is missing')
    
print(man)
print(other_man)
```

```python
# Sharpen your pencil (pg 121)
man = []
other_man = []

try:
    data = open('sketch.txt')

    for each_line in data:
        try:
            (role, line_spoken) = each_line.split(':', 1)
            if role == 'Man':
                man.append(line_spoken)
            elif role == 'Other Man':
                other_man.append(line_spoken)
        except ValueError:
            pass

    data.close()
except IOError:
    print('The data file is missing')
    
'''
This is the same as the last example, but replaces the try/except/finally pattern using
the *with* keyword; in fact, the ',' operator could also combine both *with* lines into
one line
'''
try:
    with open('man_data2.txt', 'w') as out_man:
        print(man, file=out_man)
        
    with open('other_data2.txt', 'w') as out_other:
        print(other_man, file=out_other)
    
except IOError as err:
    print('File error: '+ str(err))
```

```python
import sys

def print_lol(the_list, indent=False, level=0, fh=sys.stdout):
  """
  This function prints out a list recursively where each line will contain one
  entry from the list which may be a nested list.
  """
  for each_item in the_list:
    if isinstance(each_item, list):
      print_lol(each_item, indent, level + 1, fh)
    else:
      if indent:
        for spaces in range(level):
          print("  ", end='', file=fh)
    print(each_item, end='', file=fh)
```

```python
# Sharpen your pencil (pg 134)
import pickle

man = []
other_man = []

try:
    data = open('sketch.txt')

    for each_line in data:
        try:
            (role, line_spoken) = each_line.split(':', 1)
            if role == 'Man':
                man.append(line_spoken)
            elif role == 'Other Man':
                other_man.append(line_spoken)
        except ValueError:
            pass

    data.close()
except IOError:
    print('The data file is missing')

try:
    with open('man_data4.txt', 'wb') as out_man, open('other_data4.txt', 'wb') as out_other:
        pickle.dump(man, out_man)
        pickle.dump(other_man, out_other)
    
except IOError as err:
    print('File error: '+ str(err))
    
except pickle.PickleError as perr:
    print('Pickling error: ' + str(perr))
```
